```python
#Imports
import pandas as pd
import os, json
from pandas.io.json import json_normalize
```

```python
#Loading in the Yummly 28k dataset
path_to_json= 'metadata27638/'
json_files =[x for x in os.listdir(path_to_json) if x.endswith('.json')]
json_data = []
for json_file in json_files:
    json_file_path = os.path.join(path_to_json,json_file)
    with open(json_file_path, "r", encoding="utf8") as f:
        json_data.append(json.load(f))
```

```python
len(json_data)
```

# Preprocessing

```python
#Convert to pandas dataframe and print head, shows annoying dictionaries 
metadata = pd.DataFrame(json_data)
#Print before preprocessing
metadata.head()

#Drop irrelevant columns, extract lists of dictionaries to colums
metadata = metadata.drop(['images','attribution','source','id'],axis=1)
metadata = pd.concat([metadata.drop(['attributes'],axis=1),metadata['attributes'].apply(pd.Series)],axis=1)
metadata = pd.concat([metadata.drop(['flavors'],axis=1),metadata['flavors'].apply(pd.Series)],axis=1)
#metadata = pd.concat([metadata.drop(['nutritionEstimates'],axis=1),metadata['nutritionEstimates'].apply(pd.Series)],axis=1)
metadata = metadata[pd.notnull(metadata['Piquant'])]

#metadata = pd.concat([metadata.drop(['nutritionEstimates'],axis=1),metadata['nutritionEstimates'].apply(pd.Series)],axis=1)
#metadata = pd.concat([metadata.drop(['images'],axis=1),metadata['images'].apply(pd.Series)],axis=1)
#metadata = pd.concat([metadata.drop(['attribution'],axis=1),metadata['attribution'].apply(pd.Series)],axis=1)
```
