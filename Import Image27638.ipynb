# Import raw images and reshape them to the same size

```python
New_Size = (200,300) # size of reshaped images
```

```python
def load_image( infilename ) :
    img = Image.open( infilename )
    
    img.load()
    data = np.asarray( img, dtype="int32" )
    return data
```

```python
def load_image_resize( infilename ) :
    img = Image.open( infilename )
    img = img.resize((New_Size[1], New_Size[0]), Image.ANTIALIAS)
    img.load()
    data = np.asarray( img, dtype="int32" )
    return data
```
